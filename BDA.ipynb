```python
# Creating an RDD from a list
data = [1, 2, 3, 4, 5]
rdd = sc.parallelize(data)
print("Original RDD:", rdd.collect())

# Transformations
rdd_squared = rdd.map(lambda x: x*x)
rdd_even = rdd.filter(lambda x: x % 2 == 0)

# Actions
print("Squared Values:", rdd_squared.collect())
print("Even Values:", rdd_even.collect())
print("Sum of Values:", rdd.reduce(lambda x, y: x + y))
```
